```python
import torch
text = """
Dorothy lived in Kansas with her Aunt Em and Uncle Henry. One day, a tornado swept Dorothy and her dog Toto to the magical land of Oz. 
In Oz, Dorothy met the Scarecrow who wanted a brain, the Tin Woodman who wanted a heart, and the Cowardly Lion who wanted courage. 
They traveled together to see the Wizard of Oz in the Emerald City, hoping he would grant their wishes. 
Along the way, they faced challenges, but with friendship and courage, they overcame them. 
In the end, Dorothy discovered that she had the power to return home all along, and she woke up back in Kansas, realizing that thereâ€™s no place like home.
"""
chars = sorted(list(set(text)))
string_to_int = { ch:i for i, ch in enumerate(chars) }
int_to_string = { i:ch for i, ch in enumerate(chars) }
encode = lambda s: [string_to_int[c] for c in s]
decode = lambda l: ''.join([int_to_string[i] for i in l])
data = torch.tensor(encode(text), dtype=torch.long)
print("First 100 characters as tensor:", data[:100])
print("Vocabulary size:", len(chars))
print("Sample decoded text:", decode(data[:100].tolist()))
```
